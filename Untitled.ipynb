### Load data

```python
%matplotlib inline
import pandas as pd
import collections
from textblob import TextBlob
```

```python
df = pd.DataFrame.from_csv('tweetdata', header = 0, sep='~,~', index_col=None).dropna()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
plot = df['date'].groupby([df["date"].dt.day,df["date"].dt.hour]).count().plot(kind="bar",title='Disney Theme Parks On Twitter')
plot.set_xlabel('Day and Hour')
plot.set_ylabel('# of tweets')
```

```python
df2 = df[df['date']>'2017-07-15  21:00:00']
df2 = df2[df2['date']<'2017-07-16  8:00:00']
df2.set_index('date', drop=False, inplace=True)
plot = df2['date'].groupby(pd.TimeGrouper(freq='30Min')).count().plot(kind="bar",title='Disney Theme Parks On Twitter')
plot.set_xlabel('Hour')
plot.set_ylabel('# of tweets')
```

```python
df2 = df[df['date']>'2017-07-15  22:00:00']
df2 = df2[df2['date']<'2017-07-16  00:00:00']

# create set of words in tweets
wordlist = []
for i in df2.tweet.values:
    blob = TextBlob(i)
    wordlist += blob.noun_phrases
    #wordlist += i.replace(':','').split()
# count words
collection = collections.Counter(wordlist)
collection.most_common(n=5)
```
